# Visualizations

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create data
x = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
y1 = np.array([5.0, 4.9, 5.1, 5.4, 5.6, 5.5, 5.3, 5.2, 5.1, 4.9, 5.3, 5.5])
y2 = y1[::-1]
```

## Average Stay Graph

```python
#Create Figure
plt.style.use('dark_background')
plt.figure(dpi=200)
plt.plot(x, y1, label='National')
plt.plot(x, y2, label='Hospital')
plt.legend(loc=0)
plt.ylim([4, 6])
plt.title('Average Stay in Days')
plt.xlabel('Month')
plt.ylabel('Days')
plt.savefig('length_stay.png')
```

## Conditions Pie Chart

```python
# Create Figure
labels = 'COVID-19', 'Liveborn', 'Septicemia', 'Osteoarthritis', 'Heart failure'
sizes = [53, 21, 11, 8, 7]
explode = (0.1 , 0, 0, 0, 0) 
wedgeprops = {"edgecolor":"1",'linewidth': 1, 'antialiased': True}

fig, ax = plt.subplots(dpi=200)
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
ax.axis('equal') 
ax.set_title('Diagnoses for Inpatient Stays')
plt.tight_layout()
#Code to save as png
fig.savefig('diagnoses_pie.png', dpi=200)
```

## Bed Utilization Rate Pie Chart

```python
# Create Figure
labels = 'Occupied', 'Unoccupied'
sizes = [65, 35]
wedgeprops = {"edgecolor":"1",'linewidth': 1, 'antialiased': True}

fig, ax = plt.subplots(dpi=200)
ax.pie(sizes, colors=['coral', 'deepskyblue'], labels=labels, autopct='%1.1f%%', 
        startangle=90, wedgeprops=wedgeprops)
ax.axis('equal')  
ax.set_title('Bed Utilization Rate')
plt.tight_layout()
#Code to save as png
fig.savefig('bed_utilization_pie.png', dpi=200)
```

## Supply Left Bar Graph

```python
# Create Figure
plt.figure(dpi=200)
plt.title('Projected Days Left')
plt.ylabel('Days')
plt.xlabel('Medications')
labels = ['Tylenol', 'Cefazolin', 'Heparin', 'Zofran', 'Valium']
days_left = [5, 8, 13, 17, 18]
plt.bar(labels, days_left, edgecolor='white')
#Code to save as png
plt.savefig('medicine_supply_bargraph.png', dpi=200)
```

## Incidents Graph

```python
at_hospital = np.random.uniform(low=5, high=25, size=(12))
national_average = np.random.uniform(low=5, high=25, size=(12))

plt.figure(dpi=200)
plt.fill_between(np.arange(12), at_hospital, color="lightpink",
                 alpha=0.6, label='At Hospital')
plt.fill_between(np.arange(12), national_average, color="skyblue",
                 alpha=0.6, label='National Average')
plt.title('Number of Incidents')
plt.xlabel('Month')
plt.ylabel('Incidents')
plt.legend()
plt.savefig('incidents_area.png', dpi=200)
```

## Cost Per Dischrage Graph

```python
plt.style.use('dark_background')
plt.figure(dpi=200)
x = np.linspace(2010, 2021, 10)
y = np.array([10000, 10509, 11000, 11250, 11750, 12342, 12789, 13234, 14234, 14789])
y1 = np.array([10500, 11000, 11287, 11894, 12421, 12524, 12789, 13012, 13432, 13576])
plt.title('Avergae Cost Per Discharge')
plt.xlabel('Year')
plt.ylabel('Cost in Dollars')
plt.plot(x, y, color='lime', linewidth=2, label='United States Average')
plt.plot(x, y1, color='red', lw=2, label='Hospital')
plt.legend(loc=0)
plt.savefig('cost_per_discharge.png', dpi=200)
```
